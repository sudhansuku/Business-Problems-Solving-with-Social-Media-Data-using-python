# 2. DATA CLEANING

## a. Fixing column names

```python
content_df.columns
```

```python
# fixing column names of reactiontypes_df
reactiontypes_df = reactiontypes_df.rename(columns={"Type":"Reaction_Type"})
reactiontypes_df.columns
```

## b. fixing missing values

```python
# missing values in content_df
content_df.isnull().sum()
```

- **dropping** *URL* column as it is useless for our analysis. 

```python
# drop "URL"
content_df = content_df.drop("URL",axis=1)
content_df.head()
```

```python
# percentage of missing values in reaction_df
reaction_df.isnull().sum()/reaction_df.shape[0]*100
```

- **dropping** rows with null Reaction_Type as they are very small(3%)

```python
# dropping rows
reaction_df = reaction_df[~reaction_df.Reaction_Type.isnull()]
```

```python
# percentage of missing values in reaction_df after drop
reaction_df.isnull().sum()/reaction_df.shape[0]*100
```

```python
# looking for rows with null Reaction_User_ID   
reaction_df[reaction_df.Reaction_User_ID.isnull()].sample(10)
```

- all rows **seems ok** only Reaction_User_ID is missing completely at random, thus imputing these values with **"Unknown"**

```python
# imputing "Unknown"
reaction_df["Reaction_User_ID"] = reaction_df["Reaction_User_ID"].fillna("Unknown")

# percentage of missing values in reaction_df after drop
reaction_df.isnull().sum()/reaction_df.shape[0]*100
```

- **all missing** values are removed

```python
# missing values of Reactiontype_df
reactiontypes_df.isnull().sum()
```

- no null values

#### missing values are removed from all the datasets

## Merging all dataframes by "inner join"

```python
# merging content_df and reaction_df on "Content ID"
merged_df = pd.merge(content_df, reaction_df, on='Content ID', how='inner')

# merging merged_df and reactiontype_df on "Reaction_Type"
merged_df = pd.merge(merged_df, reactiontypes_df, on= "Reaction_Type", how="inner")

merged_df.head()
```

```python
print("rows :",merged_df.shape[0])
print("columns :",merged_df.shape[1])
```

### Percentage share of Sentiment

```python
# looking for share of posts by content_type

# Create a figure and axes.
fig, ax = plt.subplots(figsize=(7, 7))

# Plot the pie chart.
ax.pie(merged_df.Sentiment.value_counts(), labels=merged_df.Sentiment.value_counts().index, autopct="%1.1f%%",
       explode = [0.1,0,0], shadow=True)

# Customize the pie chart.
ax.set_title("Percentage share of Sentiment", fontdict={"size":20})
plt.show()
```

- Most posts are **positive (56%)** with **31% negative** and **12% neutral**

```python
merged_df.head()
```

```python
# creating dataframe of content_category aggregated by popularity score
categ_score_df = merged_df.groupby(by="Content_Category")["Score"].sum()
categ_score_df.sort_values(ascending=False,inplace=True)

# plotting bar plot showing content_category by popularity score

plt.figure(figsize=[15,7])
categ_score_df.plot.bar()
plt.title("Content_Category by Popularity Score", fontdict={"size":20})
plt.xticks(rotation=45)
plt.ylabel("total score")
plt.show()
```

- **Science & Technology** is the **most popular** category in Social Buzz.

```python
# creating dataframe of Content_type aggregated by popularity score
content_score_df = merged_df.groupby(by="Content_Type")["Score"].sum()
content_score_df.sort_values(inplace=True)

# plotting bar plot showing content_type by popularity score
plt.style.use("default")
plt.figure(figsize=[8,5])
content_score_df.plot.barh()
plt.title("Content_Type by Popularity Score", fontdict={"size":15})
plt.xticks(rotation=45)
plt.xlabel("total score")
plt.show()
```

- **Photo** is the most popular content type in Social Buzz.

```python
# creating dataframe of Content_User_ID aggregated by popularity score
categ_score_df = merged_df.groupby(by="Content_User_ID")["Score"].sum()
categ_score_df.sort_values(ascending=False,inplace=True)

# plotting bar plot showing Content_User_ID by popularity score
plt.style.use("seaborn")
plt.figure(figsize=[8,5])
sns.barplot(categ_score_df.head(10),categ_score_df.head(10).index)
plt.title("Top 10 most popular content creator", fontdict={"size":15})
plt.xticks(rotation=45)
plt.xlabel("total popularity score")
plt.ylabel("User ID")
plt.show()
```

```python
# extracting month from datetime
merged_df["month"] = merged_df["Datetime"].dt.month_name()

# merged df of top 3 categories
bool_list = [True if i in ['Science & technology', 'Education', 'Animals'] else False for i in merged_df.Content_Category ]
top3_cat = merged_df[bool_list]

# plotting bar plot showing Content_User_ID by popularity score
plt.style.use("seaborn")
plt.figure(figsize=[10,5])
sns.lineplot("month","Score", hue="Content_Category", data=top3_cat, estimator=sum)
plt.title("Poularity Trend of Top 3 category", fontdict={"size":15})
plt.xticks(rotation=45)
plt.ylabel("popularity score")
plt.show()
```

- there is **no clear trend** in the data just regular fluctuataion of popularity, with **Science & Technology** consistently popular over the year.

```python
# extracting month from datetime
merged_df["hour"] = merged_df["Datetime"].dt.hour

# plotting line plot showing most active hour
plt.style.use("seaborn")
plt.figure(figsize=[10,5])
merged_df.hour.value_counts().sort_index().plot()
plt.title("Most Active Hours", fontdict={"size":15})
plt.xticks(np.arange(0,24))
plt.xlabel("Hour")
plt.ylabel("reaction count")
plt.show()
```

- **Users** are most active in **morning** between **6am to 9am** and around **11pm at Night**

```python
# extracting weekday from datetime
merged_df["weekday"] = merged_df["Datetime"].dt.dayofweek

# plotting line plot showing most active day of the week
plt.style.use("seaborn")
plt.figure(figsize=[10,5])
merged_df.weekday.value_counts().sort_index().plot()
plt.title("Most Active weekday", fontdict={"size":15})
plt.xticks(np.arange(7),['Monday', 'Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.xlabel("weekday")
plt.ylabel("reaction count")
plt.show()
```

- **Monday** is the most active day followed by **Friday** and **Sunday**

- all content categories show **similar distribution** of **Sentiments** with slight variation.
